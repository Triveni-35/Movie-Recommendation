## observe it is a list of dictionaries 
we need genres as just list for example: ['Action','Crime','Comedy'] from every dict we need value of name 

```python
def convert(obj):
    l = []
    for i in obj:
        l.append(i['name'])
    return l
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

### OR-- if the above method raises any error
(TypeError: string indices must be integers, not 'str') due to the list was actually a string , use below method 

```python
import ast 
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(2)
```

## for cast also lets take name value of first 3 dictionaries : just assuming that first 3 dictionaries; name is important 

```python
'''import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')'''
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head(2)
```

## Still now we converted genres , keywords, cast clms
crew: we need job values as Director: Assuming director influences movie as much 

```python
# movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

## Convert the string with space to no space format
why ? 
Sam Worthington 2 tags are created for this person 
Sam mendes 2 tags are created for this person also
but Sam tag is for both model gets confused 

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ","")) 
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
'''actors' 'actress' 'acts, action these are actually similar 
so lets do stemming '''
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
print(stem('act actress action actor acting acted good'))
```

```python
## lets check this function on [0] tag first
print(stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron'))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
## We  have 4806 movies .. i.e We have 4806 vectors each of 5000 dimension 
so at higher dimension euclidian distance is not the best option 
use cosine similarity 
extract cosine similarity between every two movies among all 4806 movies 
# cosine similarity is inverse of cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)    
```

```python
recommend('Gandhi')
```
